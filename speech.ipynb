```python
import numpy as np
import librosa
import tensorflow as tf
import glob
from sklearn.model_selection import train_test_split
```

```python
max_padlen=11
aud2vec=[]
label=[]
for i in range(3):
    for filename in glob.glob("C:/Users/kumar/Desktop/DeadSimpleSpeechRecognizer-master/DeadSimpleSpeechRecognizer-master/data/"+str(i)+"/*.wav"):
        wave,sr = librosa.load(filename, mono=True, sr=None)
        wave=wave[: : 3]
        print('wave',wave.shape)
        mfcc = librosa.feature.mfcc(wave, sr=16000,n_mfcc=50)
        print('earlier_mfcc',mfcc.shape)
        pad_width = max_padlen - mfcc.shape[1]
        mfcc = np.pad(mfcc, pad_width=((0, 0), (0, pad_width)), mode='constant')
        print('after_pad',mfcc.shape)
        aud2vec.append(mfcc)
        label.append(i)
        
        
```

```python
aud2vec= np.asarray(aud2vec,dtype = np.float32)
label= np.asarray(label,dtype = np.int)
print(aud2vec.shape)
```

```python
label=np.eye(3,dtype=int)[label]
```

```python
label.shape
```

```python
X_train, X_test, y_train, y_test =train_test_split(aud2vec,label)
X_train.shape
```

```python
X_train=np.reshape(X_train,[X_train.shape[0],50,11,1])
X_test=np.reshape(X_test,[X_test.shape[0],50,11,1])
```

```python
y_train[:5]
print(X_test.shape)
```

```python
def Train(x):
    conv1=tf.layers.conv2d(x,kernel_size=5,strides=2,filters = 32, padding="same")
    conv1=tf.layers.batch_normalization(conv1)
    conv1=tf.nn.relu(conv1)
    conv1=tf.nn.dropout(conv1,0.25)
    print(conv1.shape)
    conv2 = tf.layers.conv2d(conv1,kernel_size=5,strides=2,filters = 64, padding="same")
    conv2=tf.layers.batch_normalization(conv2)
    conv2=tf.nn.relu(conv2)
    conv2=tf.nn.dropout(conv2,0.25)
    print(conv2.shape)
    conv3 = tf.layers.conv2d(conv2,kernel_size=5,strides=2,filters = 128, padding="same")
    conv3=tf.layers.batch_normalization(conv3)
    conv3=tf.nn.relu(conv3)
    conv3=tf.nn.dropout(conv3,0.25)
    print(conv3.shape)
    
    fc=tf.reshape(conv3,[-1,7*2*128])
    fc=tf.nn.relu(tf.layers.dense(fc,128,name='layer_enc_fc'))
               
    logit= tf.layers.dense(fc,3,name='layer_enc_out')
    prob=tf.nn.sigmoid(logit)
    
    return logit,prob
```

```python
tf.reset_default_graph()
batch_size=128
x=tf.placeholder('float',[None,50,11,1])
y=tf.placeholder('float',[None,3])
logit,prob=Train(x)
cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits = logit,labels=y))
optimizer=tf.train.AdamOptimizer().minimize(cost)
correct_prediction = tf.equal(tf.argmax(logit,1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
with tf.Session() as sess:
    sess.run(tf.initialize_all_variables())
    for ep in range(1000):   
            sm=[]
            for  j in range(X_train.shape[0]//128):
                    ls=X_train[j*batch_size:(j*batch_size+batch_size)]
                    otp=y_train[j*batch_size:(j*batch_size+batch_size)]
                    
#                     correct_prediction = tf.equal(tf.argmax(logit,1), tf.argmax(y, 1))
#                     accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
                    
                    _,loss,acc=sess.run([optimizer,cost,accuracy], feed_dict={x:ls,y:otp})
                    sm.append(loss)
                    
            print(ep,sum(sm)/len(sm),acc)
    valid_loss,test_acc = sess.run([cost,accuracy], feed_dict={x:X_test,y :y_test})
    print(valid_loss,test_acc)
```
